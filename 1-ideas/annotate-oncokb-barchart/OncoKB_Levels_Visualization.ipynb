OncoKB Annotator By Gene

```python
import pandas as pd
annotated_maf = pd.read_csv('example_maf.oncokb.txt', sep = "\t")
annotated_maf
```

```python
import seaborn as sns
import math
level_counts_by_gene = pd.DataFrame(columns = ['Hugo_Symbol','1_Counts','2A_Counts','2B_Counts','3A_Counts','3B_Counts','4_Counts','R1_Counts','R2_Counts','R3_Counts'])
for i, row in annotated_maf.iterrows():
    gene = list()
    for column in row["LEVEL_1":"LEVEL_R3"]:
        if(column == "NaN"):
            gene.append(0)
        else:
            drugs = column.split(",")
            count = len(drugs)
            gene.append(count)
    gene.insert(0, row['Hugo_Symbol'])
    level_count_by_gene = level_counts_by_gene.append(gene, ignore_index = True)
```
